Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY1 = """
SELECT ConvertedComp FROM master 
"""

df = pd.read_sql_query(QUERY1, conn)

df.plot(kind = 'hist', figsize = (10, 6))

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY2 = """
SELECT Age FROM master
"""

df = pd.read_sql_query(QUERY2, conn)
df.plot(kind = 'box', figsize = (10, 6))

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY3 = """
SELECT Age, WorkWeekHrs FROM master
"""

df = pd.read_sql_query(QUERY3, conn)

df.plot(kind = 'scatter', x = 'Age', y = 'WorkWeekHrs', figsize = (10, 6))

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY4 = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
"""

df = pd.read_sql_query(QUERY4, conn)

sns.scatterplot(data = df, x = 'WorkWeekHrs', y = 'CodeRevHrs', size = 'Age', sizes = (1, 500))

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY5 = """
SELECT *, COUNT(*) AS count FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY5, conn)

data = df['count']

fig, ax = plt.subplots()
ax.pie(data, labels = df['DatabaseDesireNextYear'], autopct = '%1.1f%%')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY6 = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY6, conn)

df_new = df.groupby(['Age'], as_index = False).median()
df_new.set_index('Age', inplace = True)
df_new.plot(kind = 'area', stacked = True, figsize = (10, 6))
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY7 = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY7, conn)

df = df.groupby(['Age'], as_index = True).median()
df.plot(kind = 'line', figsize = (10, 6))

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY8 = """
SELECT MainBranch, COUNT(*) AS count FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY8, conn)

sns.barplot(data = df, x = 'count', y = 'MainBranch', orient = 'h')

plt.show()
```
